# Importing Libraries

```python
import numpy as np
import pandas as pd
import csv
import datetime
from statsmodels.tsa.arima_model import ARIMA
```

# Importing Data

```python
data = pd.read_csv('Temperature-20201216-1824.csv')
```

# Viewing and Describing data

```python
data
```

```python
data.shape
```

```python
data.describe()
```

```python
temps = data['value']
```

# Defining a "temps" variable to feed the model

```python
temps
```

# Using the arima model to get the predictions

```python
model_arima=ARIMA(temps,(1,1,0))
model_arima_fit=model_arima.fit()
prediction_arima=model_arima_fit.forecast(10)[0]
prediction_arima
```

```python
prediction = pd.DataFrame(prediction_arima)
```

# Predictions and Saving to new CSV file

```python
prediction.to_csv('Predicted_Data.csv')
prediction
```

# Saving the trained model with pickle

```python
import pickle
with open('model_pickle','wb') as f:
    pickle.dump(model_arima,f)
```

# Load trained model with pickle no need to train again

```python
with open('model_pickle','rb') as f:
    df = pickle.load(f)
```
