```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
df = pd.read_csv('cluster2.csv', header = None)
df.head()
```

```python
centroids = random.sample(range(0, len(df)), 3)
centroids
```

```python
centroid = []
for i in centroids:
    centroid.append(df.loc[i])
centroid
```

```python
centroid = np.array(centroid)
X = np.array(df)
```

```python
def cal_distance(x1, x2):
    return(sum((x1 - x2)**2))**0.5
```

```python
def find_centroids(y, X):
    mark_centroid = []
    for i in X:
        distance = []
        for j in y:
            distance.append(cal_distance(i, j))
        mark_centroid.append(np.argmin(distance))
    return mark_centroid
```

```python
output_centroid = find_centroids(centroid, X)
output_centroid
```

```python
def calc_centroids(clusters, X):
    new_centroids = []
    new_df = pd.concat([pd.DataFrame(X), pd.DataFrame(clusters, columns=['cluster'])],
                      axis=1)
    for c in set(new_df['cluster']):
        current_cluster = new_df[new_df['cluster'] == c][new_df.columns[:-1]]
        cluster_mean = current_cluster.mean(axis=0)
        new_centroids.append(cluster_mean)
    return new_centroids
```

```python
for i in range(10):
    get_centroids = find_centroids(centroids, X)
    centroids = calc_centroids(get_centroids, X)
    plt.figure()
    plt.scatter(np.array(centroids)[:, 0], np.array(centroids)[:, 1], color='black')
    plt.scatter(X[:, 0], X[:, 1], alpha=0.1)
    plt.show()
    print(centroids)
```
